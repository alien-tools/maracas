# Java Libraries on GitHub

Data retrieved on 2023-08-09. We directly scrap GitHub's dependency graph webpage (e.g., https://github.com/google/guava/network/dependents). The number of clients includes every package and only count the "repositories" clients, not the "package" ones. The query is:

```lang:Java stars:>=125 pushed:>2023-02-09 is:public mirror:false template:false archived:false fork:false```

## Data

````r
repos <- read.csv('github-libraries.csv', stringsAsFactors=TRUE)

summary(repos)
````

````r
head(repos, 50)
````

## Repositories that are not libraries

We assume a number of >= 10 clients. There are many weird corner cases that can assign some clients to a repository that's not really a library.

For instance, this repository (https://github.com/akirataguchi115/akirataguchi115) declares dependencies towards tens of thousands of repositories. Forks also count as clients.

````r
libs <- subset(repos, clients >= 10)
nonlibs <- subset(repos, clients < 10)
nonlibs <- nonlibs[order(nonlibs$stars, decreasing = TRUE), ]
nrow(libs)
nrow(nonlibs)

head(nonlibs, 25)
summary(nonlibs)
````
